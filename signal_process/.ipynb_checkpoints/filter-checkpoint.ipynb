```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
sf = 30

t = np.linspace(0, 10, sf*10)
low_hz = 5
high_hz = 15
# 5 Hz
xlow = np.sin(2 * np.pi * low_hz * t)
# 250 Hz
xhigh = np.sin(2 * np.pi * high_hz * t)
x = xlow + xhigh

lowpass_hz = 10
fnorm = lowpass_hz / (0.5*sf)
b, a = signal.butter(8, fnorm, btype='highpass')

y = signal.filtfilt(b, a, x)

fig, ax = plt.subplots(2, 1, figsize=(10, 3))

ax[0].plot(x)
ax[1].plot(y)
```

```python
lowpass_hz = 10
fnorm = lowpass_hz / (0.5*250)
fnorm
```

```python
lowpass_hz = 10
fnorm = lowpass_hz / (0.5*sf)
b, a = signal.butter(8, fnorm, btype='lowpass')

y = signal.filtfilt(b, a, x)

fig, ax = plt.subplots(2, 1, figsize=(10, 3))

ax[0].plot(x)
ax[1].plot(y)
```
