```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, welch
```

```python
t = np.linspace(-5, 5, 400)

signal_num = 4

cos_3pi = np.cos(3*np.pi*t)
sin_2pi_3a = 3*np.sin(2*np.pi*t)
cos_5pi = np.cos(5*np.pi*t)
sin_4pi_2a = 2*np.sin(4*np.pi*t)

signal_lst = [cos_3pi, sin_2pi_3a, cos_5pi, sin_4pi_2a]
```

```python
fig = plt.figure(figsize=(10, (signal_num+1)*2))
ax = fig.subplots(nrows=signal_num+1, ncols=1)

for idx, signal in enumerate(signal_lst):
    ax[idx].set_ylim(-4, 4)
    ax[idx].plot(t, signal)
    
ax[-1].set_ylim(-6, 6)
ax[-1].plot(t, np.sum(signal_lst, axis=0))
```

```python
combined_signal = np.sum(signal_lst, axis=0)
```

```python
plt.plot(np.fft.fft(combined_signal))
```
