```python
from misleep import load_mat
from scipy.signal import welch
import matplotlib.pyplot as plt
```

```python
data = load_mat(r'E:\workplace\EEGProcessing\00_DATA\20240523_night\20240523_night.mat')

data.filter(['EEG'], btype='highpass', low=0.5)
data.filter(['EMG_DIFF'], btype='highpass', low=0.5)
print(data.channels)
ch_data = data.signals[2]
sf = data.sf[2]
nperseg = 5*sf

F, P = welch(ch_data, sf)

plt.plot(F, P)
plt.title("EEG 0.5 Hz highpass")
plt.show()

ch_data = data.signals[3]
sf = data.sf[3]
nperseg = 5*sf

F, P = welch(ch_data, sf)

plt.plot(F, P)
plt.title("EMG_DIFF 0.5 Hz highpass")
plt.show()
```

```python
import pandas as pd
import numpy as np
pd.DataFrame(np.array([F, P]).T, columns=['F' ,'P']).to_csv(r'E:\workplace\EEGProcessing\00_DATA\EEG_TM\20240513_190000\aaa.csv')
```
